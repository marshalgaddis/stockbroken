Ok, now we're going to use those classes to build a `Quote` object. Here's the definition of the `Quote` class:

    class Quote(object):
      
      DATE_FMT = '%Y-%m-%d'
      TIME_FMT = '%H:%M:%S'
      
      def __init__(self):
        self.symbol = ''
        self.date,self.time,self.open_,self.high,self.low,self.close,self.volume = ([] for _ in range(7))
    
      def append(self,dt,open_,high,low,close,volume):
        self.date.append(dt.date())
        self.time.append(dt.time())
        self.open_.append(float(open_))
        self.high.append(float(high))
        self.low.append(float(low))
        self.close.append(float(close))
        self.volume.append(int(volume))
          
      def to_csv(self):
        return ''.join(["{0},{1},{2},{3:.2f},{4:.2f},{5:.2f},{6:.2f},{7}\n".format(self.symbol,
                  self.date[bar].strftime('%Y-%m-%d'),self.time[bar].strftime('%H:%M:%S'),
                  self.open_[bar],self.high[bar],self.low[bar],self.close[bar],self.volume[bar]) 
                  for bar in xrange(len(self.close))])
        
      def write_csv(self,filename):
        with open(filename,'w') as f:
          f.write(self.to_csv())
            
      def read_csv(self,filename):
        self.symbol = ''
        self.date,self.time,self.open_,self.high,self.low,self.close,self.volume = ([] for _ in range(7))
        for line in open(filename,'r'):
          symbol,ds,ts,open_,high,low,close,volume = line.rstrip().split(',')
          self.symbol = symbol
          dt = datetime.datetime.strptime(ds+' '+ts,self.DATE_FMT+' '+self.TIME_FMT)
          self.append(dt,open_,high,low,close,volume)
        return True
    
      def __repr__(self):
        return self.to_csv()

    class GoogleIntradayQuote(Quote):
      ''' Intraday quotes from Google. Specify interval seconds and number of days '''
      def __init__(self,symbol,interval_seconds=300,num_days=5):
        super(GoogleIntradayQuote,self).__init__()
        self.symbol = symbol.upper()
        url_string = "http://www.google.com/finance/getprices?q={0}".format(self.symbol)
        url_string += "&i={0}&p={1}d&f=d,o,h,l,c,v".format(interval_seconds,num_days)
        csv = urllib.urlopen(url_string).readlines()
        for bar in xrange(7,len(csv)):
          if csv[bar].count(',')!=5: continue
          offset,close,high,low,open_,volume = csv[bar].split(',')
          if offset[0]=='a':
            day = float(offset[1:])
            offset = 0
          else:
            offset = float(offset)
          open_,high,low,close = [float(x) for x in [open_,high,low,close]]
          dt = datetime.datetime.fromtimestamp(day+(interval_seconds*offset))
          self.append(dt,open_,high,low,close,volume)

Breaking that down a bit:

    class Quote(object):
      
      DATE_FMT = '%Y-%m-%d'
      TIME_FMT = '%H:%M:%S'

This just declares the class that we're about to define, which we specify is a type of `object` (this is the most general predefined type of object in Python). Then we store some local variables to format dates and times. Inside of this class declaration, we're going to define some functions that will be associated with any objects we later instantiate with this class. The first function (these are also called `method`s) is special. It's called a constructor. All of the other methods defined by a class _act on_ objects of that class, but the constructor tells you how to _build_ an object of that class in the first place.

      def __init__(self):
        self.symbol = ''
        self.date,self.time,self.open_,self.high,self.low,self.close,self.volume = ([] for _ in range(7))

This says that every instance of a `Quote` object is going to have a handful of special attributes (also called `field`s). One will store its symbol, another its date, another its open value, etc. Right now we're not setting any of these values because we're not working with a specific quote yet, so they're all empty. The symbol is an empty string. The rest are empty lists. Later, when we actually instantiate this class with some real quotes, we'll fill in these values.

      def append(self,dt,open_,high,low,close,volume):
        self.date.append(dt.date())
        self.time.append(dt.time())
        self.open_.append(float(open_))
        self.high.append(float(high))
        self.low.append(float(low))
        self.close.append(float(close))
        self.volume.append(int(volume))

This just gives us an "accessor" function that will let us fill in those values safely. It just does a little preprocessing to make sure that the different attributes are of the right type (the date is a date, the open value is a real number --- a `float` --- etc.).

      def to_csv(self):
        return ''.join(["{0},{1},{2},{3:.2f},{4:.2f},{5:.2f},{6:.2f},{7}\n".format(self.symbol,
                  self.date[bar].strftime('%Y-%m-%d'),self.time[bar].strftime('%H:%M:%S'),
                  self.open_[bar],self.high[bar],self.low[bar],self.close[bar],self.volume[bar]) 
                  for bar in xrange(len(self.close))])
        
      def write_csv(self,filename):
        with open(filename,'w') as f:
          f.write(self.to_csv())

These are useful methods to have for manipulating a quote. They define functions for converting an abstract quote object to a csv file. Since they're defined in the very definition of what it means to be a `Quote` object in the context of this program, they'll automatically work for anything that we instantiate as a `Quote`.

And so on.

Then there's the `GoogleIntradayQuote` class, which we define to be a subclass of `Quote`. Since everything that's a `GoogleIntradayQuote` is also a `Quote`, it inherits all of the methods and fields associated with the latter. The `GoogleIntradayQuote` constructor here takes three parameters (ignore the `self` variable): the symbol name (`symbol`), the sampling rate for querying the data (`interval_seconds`), and the number of days over which we want to fetch data (`num_days`). The latter two parameters have defaults, if they aren't explicitly specified (the defaults are 300 and 5, respectively).

    def __init__(self,symbol,interval_seconds=300,num_days=5):

The rest of the definition of the class just tells you how to finish off this constructor. That is, right now, the only that makes a `GoogleIntradayQuote` different from a `Quote` is the fact that the former has a fancy way of getting built. Here's how we do it.

    super(GoogleIntradayQuote,self).__init__()

`super` is a keyword that accesses the methods of the superordinate class. So this just initializes all of those empty fields from `Quote`.

    self.symbol = symbol.upper()

This sets the `symbol` attribute to the value of the parameter `symbol` that was passed into the class definition.

    url_string = "http://www.google.com/finance/getprices?q={0}".format(self.symbol)
    url_string += "&i={0}&p={1}d&f=d,o,h,l,c,v".format(interval_seconds,num_days)

These lines build a local variable called `url_string` from the other parameters. Then we query Google with this url and read in the resulting webpage, which [looks like this](https://www.google.com/finance/getprices?q=ABG&i=300&p=1d&f=d,o,h,l,c,v).
The rest of this constructor definition just parses that data. It starts at the 7th line (after all that header stuff), then for every line that has exactly 5 commas (every data line), it splits the line into values, which it stores in the variables `offset`, `close`, etc. The timestamps here are a little funny. Some of them are stored as UNIX time codes (the ones starting with 'a'). The rest are just measured by their offset from that point in time. Anyway, we do a little processing, using some functions from the `datetime` library to convert the time codes to readable strings, and then we fill in all of the attributes of the `Quote` class with the `append` method at the end.

    for bar in xrange(7,len(csv)):
      if csv[bar].count(',')!=5: continue
      offset,close,high,low,open_,volume = csv[bar].split(',')
      if offset[0]=='a':

        offset = 0
      else:
        offset = float(offset)
      open_,high,low,close = [float(x) for x in [open_,high,low,close]]
      dt = datetime.datetime.fromtimestamp(day+(interval_seconds*offset))
      self.append(dt,open_,high,low,close,volume)
